```python
# apply black box blocker

# create black box blocker
bb = em.BlackBoxBlocker()
```

```python
# define a black box function. 
# The blocker function should drop tuple pairs whose last name do not match

# The function has to do the following steps

#  1) Get TrackName attributes from each of the tuples

#  2) Split TrackName attribute to get the first word (return False if eithre one is a empty string)

#  3) if last names do match return True
```

```python
def my_function(x, y):
    # x, y will be of type pandas series
    if x['TrackName']=='' or y['TrackName']=='':
        return False
    # get name attribute
    x_TrackName = x['TrackName']
    y_TrackName = y['TrackName']
    # get first word 
    x_TrackName = x_TrackName.split(' ')[0] 
    y_TrackName = y_TrackName.split(' ')[0] 
    # check if first two word match
    if x_TrackName != x_TrackName:
        return True
    else:
        return False
```

```python
# set the blackbox function
bb.set_black_box_function(my_function)
```

```python
# block tables
C3 = bb.block_tables(A, B, l_output_attrs=['TrackName'], r_output_attrs=['TrackName'])
```

```python
# display C3
C3
```

```python
# Combine all the blocker outputs
C = em.combine_blocker_outputs_via_union([C1, C2, C3])
```

```python
# display C
C
```

```python
# Use debug blocker to see if the the blocking sequence has dropped any potential matches
dbg = em.debug_blocker(C, A, B, output_size=5)
dbg
```

```python
# export C
em.to_csv_metadata(C, './candidate_set.csv')
```
