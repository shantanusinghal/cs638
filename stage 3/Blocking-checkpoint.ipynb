```python
#pd.read_csv('C:\\Users\\Zirui Tao\\Box Sync\\CS 638\\Project\\assets\\tableA.csv', nrows=15)
# read table A; table A has 'ID' as the key attribute
A = em.read_csv_metadata(PA, key = 'Id')
#em.get_key(A)
# read table B; table B has 'ID' as the key attribute
B = em.read_csv_metadata(PB, key = 'Id')
```

```python
#Blocking plan :

#A, B --attr-equiv blocking--> candset(C0) --overlap-block--> candset(C1) --overlap-block--> candset(C2) |
#                                                                                                        |----union--->candset (C) 
#A, B ---------------------------rule-based-blocking-----------------------------------> candset(C3)-----|
```

```python
# Attribute equivalence blocking: block A, B  using birth_year
ab = em.AttrEquivalenceBlocker()
C0 = ab.block_tables(A, B, 'Released', 'Released',
                    l_output_attrs=['Id', 'Artist', 'TrackName', 'Released', 'Time'], 
                    r_output_attrs=['Id', 'Artist', 'TrackName', 'Released', 'Time'])
C0
```

```python
# Overlap blocking over candidata set (C0)
ob = em.OverlapBlocker()

# block using name
C1 = ob.block_candset(C0, 'Artist', 'Artist', word_level=True, overlap_size=1)
```

```python
C1
```

```python
# Overlap blocking over candidata set (C0)
ob = em.OverlapBlocker()

# block using name
C2 = ob.block_candset(C1, 'Time', 'Time', word_level=True, overlap_size=1)
```

```python
C2
```

```python
# Rule-based blocking over input tables
# first get features that can be used
feature_table = em.get_features_for_blocking(A, B)
```

```python
len(feature_table)
```

```python
# list name related features
feature_table[feature_table.left_attribute=='TrackName']
```

```python
# create rule-based blocker
rb = em.RuleBasedBlocker()
# add rule : block tuples if name_name_lev_sim(ltuple, rtuple) < 0.7
rule = rb.add_rule(['TrackName_TrackName_lev_sim(ltuple, rtuple) < 0.7'], feature_table)
```

```python
C3 = rb.block_tables(A, B, l_output_attrs=['TrackName'], r_output_attrs=['TrackName'])
```

```python
C3
```

```python
# now debug the blocker output
dbg = em.debug_blocker(C3, A, B, output_size=10)
```

```python
dbg
```

```python
# Combine all the blocker outputs
C = em.combine_blocker_outputs_via_union([C2, C3])
```

```python
# display C
C
```

```python
# export C
em.to_csv_metadata(C, './candidate_set.csv')
```

```python
# read back C
C = em.read_csv_metadata('candidate_set.csv', ltable=A, rtable=B, encoding = 'iso-8859-1')

```
